# Question 8 Scrape data for Top 1000 Quotes of All Time


```python
driver.get("https://www.azquotes.com/")

```

```python
quotes=driver.find_element(By.XPATH, "/html/body/div[1]/div[1]/div[1]/div/div[3]/ul/li[5]/a")
quotes.click()
```

```python
quote=[]
author=[]
category=[]

for page in range(10):
    quote1=driver.find_elements(By.XPATH,'//a[@class="title"]')
    author1=driver.find_elements(By.XPATH,'//div[@class="author"]')
    category1=driver.find_elements(By.XPATH,'//div[@class="tags"]')
    for q in quote1:
        quote.append(q.text)
    for a in author1:
        author.append(a.text)
    for c in category1:
        category.append(c.text)
        
# print(len(quote),len(author),len(category))

print("Top 1000 Quotes Of All Time")
df=pd.DataFrame({"Quote":quote, "Author":author, "Type of Quotes":category})
df
```
